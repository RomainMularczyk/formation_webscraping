L'objectif de cette formation est de vous fournir les outils nécessaires pour parvenir à extraire des informations depuis des pages web : il ne s'agira pas tant de devenir un expert en développement web ou en programmation Python, que de se familiariser avec la boîte à outil de base nécessaire pour qui souhaite extraire rapidement et efficacement des informations depuis des pages web.

La formation se décompose en **trois volets** (1. HTML, 2. Python, 3. Web scraping), dont deux préalables nécessaires (1. et 2.) pour parvenir à l'objectif final : récupérer automatiquement des grands lots de données variées (télécharger des fichiers, récupérer des métadonnées bibliographiques, constituer de grands corpus de textes) depuis des pages web. On ne traitera ainsi que la partie en amont du travail de recherche : la collecte et la structuration de données en vue de leur analyse.

Si elle a une finalité pratique et technique, cette formation constitue également un élément de culture numérique et vise, en ce sens, à interroger des outils que nous utilisons au quotidien (navigateur web, en particulier), d'en ouvrir les boîtes noires qu'ils peuvent constituer, afin de mieux les comprendre pour savoir les mobiliser pour répondre à nos questions de recherche (mais également en saisir les limites et les contraintes). Je vous renvoie ici, en particulier, aux travaux de Jack Goody (sur la question de la "literacy") pour interroger nos rapports contemporains à l'écriture et aux cultures de l'écrit. La collecte, la structuration et la manipulation des données numériques constitue un élément de cette culture de l'écrit : elle s'appuie sur des règles particulières, mais également des tours de mains, des façons de faire et des habitudes particulières qui s'ancrent avant tout par la pratique.

Ainsi, il s'agira avant tout de retrousser ses manches et de commencer à dialoguer avec notre nouveau compagnon de route : le clavier.

## <div id="HTML">1. HTML</div>

### <div id="Eléments d'introduction">1. 1. Eléments d'introduction</div>

Le <strong>langage HTML</strong> fait partie de la famille des <strong>langages à balises</strong> : c'est un <strong>langage de description</strong>, c'est-à-dire qu'il permet de structurer des données selon une série de règles et de contraintes spécifiquement définies. Il existe d'autres langages de description, les plus courants aujourd'hui étant XML et JSON (les éléments que nous apprendrons durant cette formation sont mobilisables pour traiter des fichiers XML).
<br>
<br>
Un fichier HTML n'est rien d'autre qu'une succession de <strong>balises</strong>. Les balises sont les atomes, les éléments de base des langages à balises. Une balise s'écrit à l'aide de <strong>chevrons</strong>. Pour illustrer le propos, prenons l'exemple d'un objet concret : un livre. Nous tenterons de le décrire en utilisant des balises.
<br>
<br>
La balise n'est pas un élément unique mais fonctionne en couple. En effet, une balise comporte en réalité deux éléments écrits : une <strong>balise ouvrante</strong> `<livre>` et une <strong>balise fermante</strong> `</livre>`. La balise ouvrante contient donc une chaîne de caractères quelconque encadrée par un <strong>chevron ouvrant</strong> `<` et un <strong>chevron fermant</strong> `>`. La <strong>balise fermante</strong> s'écrit de la même façon, mais le chevron ouvrant est directement suivi d'un <strong>slash</strong> `/`.

A elles deux, elles permettent d'encapsuler, d'enfermer les sous-éléments que contiennent un objet donné (dans notre cas, le livre, contiendra d'autres informations : son titre, son auteur, sa date de publication, etc.). On peut ainsi combiner et imbriquer les balises pour décrire notre livre :
```html
<livre>
    <titre>L'ingénu</titre>
    <auteur>Voltaire</auteur>
    <date>1767</date>
</livre>
```

#### <div id="Que peuvent contenir les balises ?">1. 1. 1. Que peuvent contenir les balises ?</div>

Une balise peut contenir différents éléments :
* Une valeur (une chaîne de caractères, un nombre)
* Une autre balise
* Un mélange des deux

```html
<livre>
    <titre>L'ingénu</titre>
    <auteur>Voltaire
        <nom>François-Marie Arouet</nom>
    </auteur>
    <date>1767</date>
</livre>
```

Si une balise ne contient aucune autre information, on peut alors l'écrire sous la forme d'une <strong>balise auto-fermante</strong> : `<livre/>`.

#### <div id="Les attributs">1. 1. 2. Les attributs</div>

Il est possible d'associer des informations complémentaires à une balise. Pour ce faire, on leur adjoint ce que l'on nomme un <strong>attribut</strong>. Il donne généralement une information complémentaire sur le contenu de la balise mais ne représente pas, à proprement parler, le contenu même de cette balise : il est une information supplémentaire attaché, qui fait partie même de la balise.
<br>
<br>
Tout comme pour une balise, on peut lui attribuer n'importe quel nom. L'attribut est renseigné entre les deux chevrons de la balise ouvrante, de la sorte :
```html
<livre id="version illustrée">
    <titre>L'ingénu</titre>
    <auteur>
        <nom class="pseudonyme">Voltaire</nom>
        <nom class="nom de naissance">François-Marie Arouet</nom>
    </auteur>
    <date>1767</date>
</livre>
```

#### <div id="Les commentaires">1. 1. 3. Les commentaires</div>

Dernier élément : il est possible d'ajouter des commentaires dans un fichier HTML. Il s'agit de portions de texte qui ne seront pas interprétées à la lecture du fichier par l'ordinateur. Pour insérer un commentaire, on utilise la syntaxe suivante : `<!-- Insérez ici votre commentaire -->`.

### 1. 2. Le cas spécifique du HTML, son environnement technique

Tous les éléments que nous avons vu précédemment sont autant valables pour du XML que du HTML. Cependant, les documents HTML n'utilisent qu'un nombre limité et précisément défini de <strong>balises</strong> et d'<strong>attributs</strong>. Il ne s'agira pas de tous les décrire ici, mais de les définir au fil de l'eau en travaillant sur des exemples concrets.

____

Pour nous guider, nous pourrons utiliser la documentation officielle d'HTML 5 (la version actuelle du langage) : https://dev.w3.org/html5/html-author/.

Pour une ressource plus "accueillante", on pourra se référer au site MDN web docs : https://developer.mozilla.org/fr/docs/Web/HTML/Element.

#### 1. 2. 1. Le web et l'environnement "client-serveur"

S'il ne s'agit ici, ni de faire une histoire du web, ni une plongée en profondeur dans les fonctionnements techniques du web et d'internet, il est tout de même utile de disposer d'une connaissance minimale sur ce plan. Pour la suite de notre formation, il sera notamment utile de faire la distinction entre une <strong>page web dynamique</strong> et une <strong>page web statique</strong>. Pour ce faire, il s'agit avant tout de comprendre la logique de fonctionnement <strong>"client-serveur"</strong>.



<img src='img/client-serveur.png' height='500' width='500'/>

Il s'agit d'un mode de communication entre ordinateurs **"demandeurs" d'information** et ordinateurs **"fournisseurs" d'information**. Les micro-ordinateurs personnels que nous utilisons jouent ainsi (la plupart du temps), le rôle de "clients" : ils envoient une **requête** à un "serveur", lequel est en mesure de la comprendre et d'envoyer une **réponse** au "client", souvent, une page web écrite en langage HTML (dans le contexte du web).

C'est par le biais de ces **requêtes** que l'on peut obtenir des pages web exploitables à partir desquelles on peut sélectionner et extraire les informations qui nous intéressent.

#### 1. 2. 2. Les balises génériques

Un fichier HTML contient toujours un certain nombre de balises élémentaires. Celles-ci permettent notamment d'indiquer toute une série d'informations fonctionnelles qui permettent, entre autre, d'appliquer des éléments de style au contenu. En effet, les langages de description ne sont pas véritablement conçus pour formater et éditer le rendu visuel du contenu : certaines balises permettent ainsi de spécifier comment accéder à d'autres fichiers utiles à l'affichage de la page web.

On peut ainsi se représenter les balises HTML comme des jalons, des prises auxquelles on peut "s'accrocher" pour positionner un curseur (si on fait l'analogie de la façon dont on utilise la souris pour sélectionner du texte dans un document) au sein d'un document HTML. Comme les balises HTML contiennent des informations, on peut ainsi les utiliser pour filtrer et sélectionner les éléments que l'on souhaite retenir.

Si le langage HTML ne permet pas de définir le rendu visuel de contenu, il permet néanmoins de le structurer.

|Nom de la balise|Format|Arborescence|Spécificité|
|---|---|---|---|
|`<div>`|Balise ou texte|Unique|Aucune|
|`<span>`|Texte seulement|Unique|Aucune|
|`<ul>`|Liste|Parent|Liste non-ordonnée|
|`<ol>`|Liste|Parent|Liste ordonnée|
|`<li>`|Liste|Enfant|   |
|`<a>`|Lien hypertexte|Unique|   |

#### 1. 2. 3. Le contenu dynamique

Pour collecter des données sur le web, il est important de pouvoir distinguer les pages statiques des pages dynamiques. Ces dernières sont en effet particulière dans le sens où elles n'ont pas d'existence "à part entière" sur le web : elles sont générées à la suite d'une requête effectuée par un "client". On ne peut pas véritablement y accéder, il s'agit plutôt de pages que l'on "demande".

Un exemple très simple pour illustrer la différence entre page web statique et page web dynamique est l'usage que vous pouvez faire de sites internet comme https://www.oui.sncf/. C'est seulement une fois que vous avez renseigné vos critères de recherche pour trouver votre billet de train que le serveur vous retournera une page contenant les résultats qui correspondent à votre recherche. On accède pas véritablement à la page, on la demande au serveur.

## <div id="Le DOM">2. Le DOM</div>

### <div id="Définition">2. 1. Définition</div>

Le **DOM (pour Document object model)** est, comme son nom l'indique, un modèle qui vise à représenter le document affiché par le navigateur. Il permet à des programmes d'accéder au document et d'en modifier le contenu. Dans notre cas, on pourra le voir comme une forme de voie d'escalade nous offrant des prises pour "accrocher" les éléments HTML que l'on souhaitera extraire du site.

Contrairement aux balises qui ne permettent d'identifier que des "blocs d'information", le **DOM** permet de représenter également le contenu de ces balises, qu'il soit constitué d'**attributs** ou de **plein texte**. Il représente la **structure arborescente** du document HTML, décomposant ce dernier en **noeuds**, à partir desquels se développent des **branches**.

### <div id="Les événements">2. 2. Les événements</div>

Au-delà du fait de représenter un document HTML et de le rendre modifiable par le biais de langages de programmation, il permet également de surveiller des **événements**. Par ce terme sans doute un peu trompeur, on désigne des **actions réalisées par l'utilisateur** qui navigue sur un site web. Si ces actions sont nombreuses (changement de taille de la fenêtre du navigateur, clic de souris, touche de clavier enfoncée ou relachée, etc.), celles qui nous intéresserons le plus seront avant tout les clics souris puisqu'ils peuvent permettre de naviguer d'une page à une autre.

Dans certains cas et en particulier dans le cas du contenu généré de façon dynamique, il pourra être utile d'utiliser les **événements du DOM** pour pouvoir parcourir le contenu d'une page web.

## <div id="Les requêtes HTTP">2. Les requêtes HTTP</div>

### <div id="Définition et exemples">2. 1. Définition et exemples</div>

**HTTP (pour Hyptertext Transfer Protocol)** est un protocole de transfert d'information client-serveur utilisé pour envoyer des **pages HTML** : si on se rappelle de la définition donnée plus tôt, il s'agit d'une façon pour des **machines clientes** de communiquer avec des **serveurs**. Les **navigateurs** que nous utilisons sont des **clients HTTP** qui nous permettent d'interroger des serveurs distants via l'interface que constituent les pages web.

Une **requête HTTP** se présente sous la forme suivante :
* Une ligne de commande, suivie de l'URL demandée, suivie de la version du protocole HTTP
* Une en-tête de requête
* Une ligne vide
* Le corps de la requête (Par exemple, lorsqu'on envoie des informations pour effectuer une recherche)

Un exemple ci-dessous :

```
GET https://en.wikipedia.org/wiki/Michael_Jackson HTTP/1.0
User-Agent: Mozilla
Accept: */*
```

L'important ici n'est pas tant d'apprendre à écrire des requêtes HTTP que de comprendre la logique globale de l'échange d'information entre clients et serveurs par ce biais. Pour cela, il est par contre utile de savoir lire des **statuts de requêtes** et de faire la distinction entre les **différentes commandes** que l'on peut envoyer à l'aide d'une requête HTTP.

Ci-dessous, une liste des commandes les plus courantes :

|Commande|Description|Effet sur la ressource|
|---|---|---|
|`GET`|Il s'agit de la requête que l'on utilise dans le cas général. Il s'agit d'une simple demande de page.|Sans|
|`POST`|Il s'agit de la deuxième requête la plus fréquente. Elle permet de transmettre des données entrées par un client.|Avec|
|`PUT`|Elle permet d'ajouter ou de modifier une ressource sur le serveur. Dans les faits, elle n'est quasiment jamais utilisée et on préfère toujours utiliser la méthode `POST` pour modifier du contenu sur une page.|Avec|
|`DELETE`|Elle permet de supprimer une ressource sur le serveur. De la même manière que pour la requête `PUT`, celle-ci est rarement utilisée, au bénéfice du `POST`.|Avec|
